```python
# Imported required libraries

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.impute import SimpleImputer
```

```python
# Loading the dataset
df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
```

```python
#I inspected the data first

print(df.shape)
df.head()
```

```python
# Continuous target for Linear Regression

y_house_price = df['SalePrice']
```

```python
# Binary target for Logistic Regression

# If house price is greater than 200,000 then BuyHouse = 1 else 0

df['BuyHouse'] = (df['SalePrice'] > 200000).astype(int)
y_buy_house = df['BuyHouse']
```

```python
features = ['LotArea', 'YearBuilt', 'OverallQual', 'GrLivArea', 'FullBath', 'TotRmsAbvGrd']
X = df[features]
```

```python
# Train Test Split , did single split for both targets
X_train, X_test, y_price_train, y_price_test, y_buy_train, y_buy_test = train_test_split(
    X, y_house_price, y_buy_house, test_size=0.2, random_state=42
)
```

```python
# Handled missing values (avoid data leakage)
imputer = SimpleImputer(strategy='mean')

X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=features)
X_test = pd.DataFrame(imputer.transform(X_test), columns=features)
```

```python
# Did Feature scaling
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Preprocessing complete.")
```

```python
# Linear Regression – House Price

linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_price_train)

price_predictions = linear_model.predict(X_test_scaled)

rmse = np.sqrt(mean_squared_error(y_price_test, price_predictions))
print(f"Linear Regression RMSE: ${rmse:,.2f}")
```

```python
# Logistic Regression – BuyHouse

logistic_model = LogisticRegression()
logistic_model.fit(X_train_scaled, y_buy_train)

buy_predictions = logistic_model.predict(X_test_scaled)

accuracy = accuracy_score(y_buy_test, buy_predictions)
print(f"Logistic Regression Accuracy: {accuracy:.2%}")
```
