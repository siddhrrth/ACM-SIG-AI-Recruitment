```python
# I imported the required libraries for data analysis and visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.set(style="whitegrid")
```

```python
# I loaded the Students Performance in Exams dataset from Kaggle

df = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
# I examined summary statistics of numerical features

df.describe()
```

```python
# I checked for missing values to identify any data quality issues

df.isnull().sum()
```

```python
plt.figure()
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

```python
plt.figure()
sns.histplot(df['reading score'], kde=True)
plt.title("Distribution of Reading Scores")
plt.show()
```

```python
plt.figure()
sns.histplot(df['writing score'], kde=True)
plt.title("Distribution of Writing Scores")
plt.show()
```

```python
# I compared math score distributions across genders

plt.figure()
sns.boxplot(x='gender', y='math score', data=df )
plt.title("Math Score Distribution by Gender")
plt.show()
```

```python
# Likewise previous comparison did rest of comparisons

plt.figure()
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title("Math Scores vs Test Preparation Course")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='lunch',y='math score', data=df)
plt.title("Math Scores by Lunch Type")
plt.show()
```

```python
# I calculated and visualized correlations between math, reading, and writing scores


plt.figure()
sns.heatmap(
    df[['math score','reading score','writing score']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Subject Scores")
plt.show()
```

I observed that the dataset didn't contain missing values. However, a few outliers were present in subject score distributions, which may slightly influence the analysis.

**The analysis shows that math, reading, and writing scores are strongly correlated. Test preparation and lunch type have a noticeable impact on student performance, while minor score variations are observed across demographic groups.**
